```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca dataset
df = pd.read_csv("electronic_shop.csv")
df
```

```python
# Konversi kolom transaction_date ke dalam format datetime
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Membuat line plot untuk jumlah nilai transaksi per hari
plt.figure(figsize=(10, 6))
df.groupby(df['transaction_date'].dt.date)['transaction_amount'].sum().plot(marker='o', linestyle='-')
plt.title('Jumlah Nilai Transaksi per Hari', fontsize=16)
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Jumlah Nilai Transaksi', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Konversi kolom transaction_date ke dalam format datetime
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Membuat bar plot untuk jumlah customer per hari
plt.figure(figsize=(10, 6))
df.groupby(df['transaction_date'].dt.date)['customer_name'].nunique().plot(kind='bar', color='skyblue')
plt.title('Jumlah Customer per Hari', fontsize=16)
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Jumlah Customer', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Membuat scatter plot untuk jumlah nilai transaksi per hari
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.groupby(df['transaction_date'].dt.date)['transaction_amount'].sum().reset_index(), x='transaction_date', y='transaction_amount', color='salmon')
plt.title('Scatter Plot Jumlah Nilai Transaksi per Hari', fontsize=16)
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Jumlah Nilai Transaksi', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
