```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Baca data dari file CSV
df = pd.read_csv("survey.csv")
print(df)
```

```python
# Preprocessing data untuk menghindari duplikat dan data dengan atribut yang tidak lengkap
required_columns = ['name','age', 'favorite_front_end', 'favorite_back_end', 'salary_per_year_in_USD']
df.drop_duplicates(inplace=True)
df.dropna(subset=required_columns, inplace=True)
```

```python
# Persentase responden berdasarkan umur
plt.figure(figsize=(8, 6))
df['age'].value_counts(normalize=True).sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Persentase Responden Berdasarkan Umur')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan front end
plt.figure(figsize=(8, 6))
df['favorite_front_end'].value_counts().plot(kind='line', marker='o')
plt.title('Bahasa Pemrograman/Framework Front End yang Paling Banyak Digunakan')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Jumlah Responden')
plt.grid(True)  # Menambahkan grid pada plot
plt.tight_layout()
plt.show()
```

```python
# Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan back end
plt.figure(figsize=(8, 6))
backend_counts = df['favorite_back_end'].value_counts()
colors = range(len(backend_counts))  # Menghasilkan warna yang berbeda untuk setiap titik
plt.scatter(backend_counts.index, backend_counts.values, c=colors, s=100, alpha=0.5)
plt.title('Bahasa Pemrograman/Framework Back End yang Paling Banyak Digunakan')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Jumlah Responden')
plt.tight_layout()
plt.show()
```

```python
# Persentase bahasa pemrograman / framework yang dipilih dalam pengembangan front end
plt.figure(figsize=(8, 6))
frontend_percentages = df['favorite_front_end'].value_counts(normalize=True) * 100
plt.pie(frontend_percentages, labels=frontend_percentages.index, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Bahasa Pemrograman/Framework Front End yang Dipilih')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['salary_per_year_in_USD'], bins=10, edgecolor='black')
plt.title('Distribusi Gaji Per Tahun')
plt.xlabel('Gaji Per Tahun (USD)')
plt.ylabel('Frekuensi')
plt.tight_layout()
plt.show()
```
