```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca dataset dengan menyesuaikan format tanggal
df = pd.read_csv("bakery.csv", parse_dates=['transaction_date'], dayfirst=True)
df
```

```python
# Jumlah nilai transaksi per hari
daily_transaction_sum = df.groupby(df['transaction_date'].dt.date)['transaction_amount'].sum()

# Visualisasi 
plt.figure(figsize=(12, 6))
sns.barplot(x=daily_transaction_sum.index, y=daily_transaction_sum.values, color='skyblue')
plt.title('Jumlah Nilai Transaksi per Hari', fontsize=16)
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Jumlah Nilai Transaksi', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Jenis roti yang paling banyak dipesan
most_ordered_bread = df['item_name'].value_counts().idxmax()

# Visualisasi 
plt.figure(figsize=(8, 8))
df['item_name'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Jenis Roti yang Paling Banyak Dipesan', fontsize=16)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Menghitung rata-rata kepuasan pelanggan per hari
daily_customer_rating_avg = df.groupby('transaction_date')['customer_rating'].mean()

# Visualisasi: Line plot untuk rata-rata kepuasan pelanggan per hari
plt.figure(figsize=(15, 7))
plt.plot(daily_customer_rating_avg.index, daily_customer_rating_avg.values)

plt.xlabel('Transaction Date')
plt.ylabel('Average Customer Rating')
plt.title('Average Customer Rating per Day')

plt.xticks(rotation=50)

plt.show()
plt.show()
```
